```python
from utils.file_system_reader import FileSystemReader
```

```python
reader = FileSystemReader("read_json_multi.json")
data = reader.read_file()
```

check if it is a single passage of multi

```python
if isinstance(data,dict):
    print("Single Passage Detected")
```

```python
multi_stories =[]
if isinstance(data,list):
    print("Multi Passage Detected ! Get ready for a big loop")
    multi_stories = data

print(multi_stories)
```

```python
from services.clean_tokenized_english_words_array import CleanEnglishTokenizedData
from services.extract_definitions_english_words import ExtractDefinitions
from services.extract_english_synonym_antonym import ExtractEnglishSynonymAntonym
from services.tokenize_english_passage import TokenizeEnglishVersion
from services.tokenize_sanskrit_passage_web import TokenizeSanskritPassageWeb
```

```python
normalized_passages = []

for item in multi_stories:
    normalized_passages.append({
        "title": {
            "englishVersion": item["englishTitle"],
            "sanskritVersion": item["sanskritTitle"]
        },
        "englishVersion": item["englishPassage"],
        "sanskritVersion": item["sanskritPassage"]
    })

print(normalized_passages)
```

```python
tokenized_data = []
for passage in normalized_passages:
    eng_tokenizer = TokenizeEnglishVersion(passage)
    eng_tokenized = eng_tokenizer.tokenize_english_version()
    eng_synonymzer = ExtractEnglishSynonymAntonym(eng_tokenized)
    eng_synonym_added = eng_synonymzer.execute()
    def_adder = ExtractDefinitions(eng_synonym_added)
    eng_def_added = def_adder.execute()
    eng_cleaner = CleanEnglishTokenizedData(eng_def_added)
    eng_cleaned_data = eng_cleaner.execute()
    sa_tokenizer = TokenizeSanskritPassageWeb(eng_cleaned_data)
    final_data = sa_tokenizer.tokenize()
    tokenized_data.append(final_data)

print(tokenized_data)
    
```
