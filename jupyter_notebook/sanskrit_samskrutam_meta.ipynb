```python
import requests
from uuid import uuid4
```

```python
url_story1 = "https://sanskrit.samskrutam.com/en.literature-stories-01.ashx"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "text/html,application/xhtml+xml",
    "Referer": "https://sanskrit.samskrutam.com/"
}

session = requests.Session()
session.headers.update(headers)

r = session.get(url_story1)

RAW_DATA = r.text
```

```python
print(RAW_DATA)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(RAW_DATA,"html.parser")
REQUIRED_DATA = soup.find("div",id="PageContentDiv")
print(REQUIRED_DATA)
```

```python
anchors = soup.find_all("a", attrs={"name": True})
print(anchors)
```

```python
vendor_id = []
names = [a["name"] for a in soup.find_all("a", attrs={"name": True})]
print(names)
```

```python
titles = soup.find_all("span",class_ ="contentBold3")
print(titles)
```

```python
import re

result = []

for tag in soup.find_all("span", class_="contentBold3"):
    text = tag.get_text(" ", strip=True)

    if " -- " in text:
        sanskrit, english = text.split(" -- ", 1)
    elif " - " in text:
        sanskrit, english = text.split(" - ", 1)
    else:
        sanskrit, english = text, ""

    result.append({
        "sanskrit_version": sanskrit.strip(),
        "english_version": english.strip()
    })

print(result)
```

Fetching Single Story

```python
stories = soup.find_all("span", class_="contentBold3")
anchors = soup.find_all("a", attrs={"name": True})
print(stories)
print(anchors)
```

```python
import re
from uuid import uuid4

META_DATA = []

stories = soup.find_all("span", class_="contentBold3")
anchors = soup.find_all("a", attrs={"name": True})

block = {
    "source": "story1",
    "source_url": url_story1,
    "stories": []
}

for anchor, span in zip(anchors, stories):
    raw = span.get_text(" ", strip=True)

    # Split only on the separator between Sanskrit and English
    if " -- " in raw:
        sanskrit_title, english_title = raw.split(" -- ", 1)
    elif " - " in raw:
        sanskrit_title, english_title = raw.split(" - ", 1)
    else:
        sanskrit_title = raw
        english_title = ""

    block["stories"].append({
        "_id": str(uuid4()),
        "vendorId": anchor.get("name"),
        "sanskrit_title": sanskrit_title.strip(),
        "english_title": english_title.strip(),
    })

META_DATA.append(block)

print(META_DATA)
```
