Reading From File System

```python
from utils.file_system_reader import FileSystemReader
FILE_NAME = "stories_raw.json"
```

```python
reader = FileSystemReader(FILE_NAME)
RAW_DATA = reader.read_file()
print(RAW_DATA)
```

Tokenize English Words

```python
from services.fable_extraction_pipeline.tokenize_english_version_story import TokenizeEnglishVersion
```

```python
tok = TokenizeEnglishVersion(RAW_DATA)
TOKENIZED_ENGLISH_VERSION = tok.tokenize_english_version()
print(TOKENIZED_ENGLISH_VERSION)
```

Adding synonyms and antonyms to tokenized english words

```python
from services.fable_extraction_pipeline.extract_english_synonym_antonym import ExtractEnglishSynonymAntonym
```

```python
diction = ExtractEnglishSynonymAntonym(TOKENIZED_ENGLISH_VERSION)
GRAMMAR_ADDED_VERSION = diction.execute()
print(GRAMMAR_ADDED_VERSION)
```

Tokenize Sanskrit version

```python
from services.fable_extraction_pipeline.tokenize_sanskrit_version_story import TokenizeSanskritVersion
```

```python
req = TokenizeSanskritVersion(GRAMMAR_ADDED_VERSION)
PRE_CLEANED_DATA = req.tokenize_sanskrit()
print(PRE_CLEANED_DATA)
```

Cleaning data before storage

```python
from services.fable_extraction_pipeline.clean_english_tokenized import CleanEnglishTokenizedData
```

```python
req = CleanEnglishTokenizedData(PRE_CLEANED_DATA)
CLEANED_DATA = req.execute()
print(CLEANED_DATA)
```
