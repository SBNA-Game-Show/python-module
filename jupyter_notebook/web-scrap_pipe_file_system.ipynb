Reading From File System

```python
from utils.file_system_reader import FileSystemReader
FILE_NAME = "stories_raw.json"
```

```python
reader = FileSystemReader(FILE_NAME)
RAW_DATA = reader.read_file()
print(RAW_DATA)
```

Tokenizing English Passage

```python
from services.tokenize_english_passage import TokenizeEnglishVersion
```

```python
tokenize_english_passage = TokenizeEnglishVersion(RAW_DATA)
tokenized_english_passage = tokenize_english_passage.tokenize_english_version()
print(tokenized_english_passage)
```

Add Synonyms and Antonyms to Tokenized English words

```python
from services.extract_english_synonym_antonym import ExtractEnglishSynonymAntonym
```

```python
synonyms_english_words = ExtractEnglishSynonymAntonym(tokenized_english_passage)
synonyms_antonyms_eng_words = synonyms_english_words.execute()
print(synonyms_antonyms_eng_words)
```

Tokenizing Sanskrit Passage

```python
from services.tokenize_sanskrit_passage import TokenizeSanskritVersion
```

```python
req = TokenizeSanskritVersion(RAW_DATA)
tokenized_sanskrit = req.tokenize_sanskrit()
print(tokenized_sanskrit)
```

Cleaning Tokenized English Words for Duplicates and other

```python
from services.clean_tokenized_english_words_array import CleanEnglishTokenizedData
```

```python
uncleaned_data = CleanEnglishTokenizedData(synonyms_antonyms_eng_words)
CLEANED_DATA = uncleaned_data.execute()
print(CLEANED_DATA)
```
