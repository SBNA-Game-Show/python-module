```python
from utils.file_system_reader import FileSystemReader
```

```python
request = FileSystemReader("stories_raw.json")
RAW_DATA = request.read_file()
print(RAW_DATA)
```

```python
SANSKRIT_VERSION = RAW_DATA["sanskritVersion"]
print(SANSKRIT_VERSION)
```

```python
SANSKRIT_VERSION_SENTENCE_ARRAY = []

for sentence in SANSKRIT_VERSION:
    SANSKRIT_VERSION_SENTENCE_ARRAY.append(sentence.split())

print(SANSKRIT_VERSION_SENTENCE_ARRAY)
```

```python
SANSKRIT_WORD_ARRAY =[]
for item in SANSKRIT_VERSION_SENTENCE_ARRAY:
    for word in item:
        SANSKRIT_WORD_ARRAY.append(word)

print(SANSKRIT_WORD_ARRAY)
```

```python
SANSKRIT_TOKENIZED_WORDS = []

for word in SANSKRIT_WORD_ARRAY:
    doc = nlp(word)
    stanza_word = doc.sentences[0].words[0]
    
    SANSKRIT_TOKENIZED_WORDS.append({
        'word': stanza_word.text,
        'upos': stanza_word.upos,
        'xpos': stanza_word.xpos,
        'feats': stanza_word.feats if stanza_word.feats else "_"
    })

print(SANSKRIT_TOKENIZED_WORDS) 
```

```python
full_text = " ".join(SANSKRIT_WORD_ARRAY)

# 2. Run the NLP pipeline once on the full context
doc = nlp(full_text)

SANSKRIT_TOKENIZED_WORDS = []

# 3. Iterate through sentences and words in the processed document
for sentence in doc.sentences:
    for stanza_word in sentence.words:
        SANSKRIT_TOKENIZED_WORDS.append({
            'word': stanza_word.text,
            'upos': stanza_word.upos,
            'xpos': stanza_word.xpos,
            'feats': stanza_word.feats if stanza_word.feats else "_"
        })

# To see the vastly improved outcome:
print(SANSKRIT_TOKENIZED_WORDS) # Printing the first 5 to verify
```
