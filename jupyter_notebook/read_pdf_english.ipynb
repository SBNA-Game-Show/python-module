Reading Single English Passage from PDF 

```python
import os
import pymupdf
```

```python
file_path = os.path.join("..","data","three_little_pigs.pdf")
print(file_path)
```

```python
doc = pymupdf.open(file_path)

english_lines = []

for page in doc:
    data = page.get_text("dict")

    for block in data["blocks"]:
        if "lines" not in block:
            continue

        block_text = []

        for line in block["lines"]:
            line_text = "".join(span["text"] for span in line["spans"]).strip()
            if line_text:
                block_text.append(line_text)

        if block_text:
            english_lines.append(" ".join(block_text))

print("\n\n".join(english_lines))
```

```python
print((english_lines[1]))
```

```python
def split_title_body(lines):
    title = lines[0] if lines else ""
    body = "\n".join(lines[1:]) if len(lines) > 1 else ""
    return title, body


title, body = split_title_body(english_lines)

english_object = {
    "title": title,
    "englishVersion": body
}

print(english_object)
```

Reading Multiple English Passages from PDF

```python
import pymupdf

passage_array = []
current_title = None
current_text = []

doc = pymupdf.open(file_path)

for page in doc:
    text = page.get_text()

    for line in text.splitlines():
        line = line.strip()

        if not line:
            continue

        # Detect title
        if line.isupper():
            # Save previous passage
            if current_title:
                passage_array.append({
                    "title": current_title,
                    "englishVersion": " ".join(current_text).strip()
                })

            current_title = line
            current_text = []

        else:
            current_text.append(line)

# Save last passage
if current_title:
    passage_array.append({
        "title": current_title,
        "englishVersion": " ".join(current_text).strip()
    })

print(passage_array)
```
