```python
reader = FileSystemReader("stories_raw.json")
raw_data = reader.read_file()
print(raw_data)
```

```python
sanskrit_passage = raw_data["sanskritVersion"]
print(sanskrit_passage)
```

Transliterating sanskrit passage

```python
transliterated_passage = []

for sentence in sanskrit_passage:
    transliterated_version = sanscript.transliterate(
        sentence,
        sanscript.DEVANAGARI,
        sanscript.OPTITRANS_DRAVIDIAN
    )
    transliterated_passage.append(transliterated_version)

print(transliterated_passage)
```

Cleaning Passage

```python
cleaned_transliterated_passage = []

for sentence in transliterated_passage:
    sentence = sentence.replace("|", "")
    sentence = sentence.replace("||", "")
    sentence = sentence.replace("-", "")
    sentence = sentence.replace("–", "")
    sentence = sentence.replace("—", "")
    sentence = sentence.replace('"', "")
    sentence = sentence.replace("“", "")
    sentence = sentence.replace("”", "")
    sentence = sentence.replace(",", "")
    sentence = sentence.replace(".", "")
    sentence = sentence.replace("?", "")
    sentence = sentence.replace("!", "")
    sentence = sentence.replace(";", "")
    sentence = sentence.replace(":", "")
    sentence = sentence.replace("(", "")
    sentence = sentence.replace(")", "")

    cleaned_transliterated_passage.append(sentence)
print(cleaned_transliterated_passage)
```

Importing Model

```python
model_name = "chronbmm/sanskrit5-multitask"
max_length = 512
model = T5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
def read_skt_tags(path):
    result = {}
    lines = open(path).readlines()
    lines = [line.split("\t") for line in lines]
    for line in lines:
        result[line[0]] = line[1].strip()
    return result

sanskrit_tags = read_skt_tags("../data/sanskrit_tags.tsv")


def process_batch(batch, mode):
    prefix = {
        'segmentation': "S ",
        'segmentation-morphosyntax': "SM ",
        'lemma': "L ",
        'lemma-morphosyntax': "LM ",
        'segmentation-lemma-morphosyntax': "SLM "
    }

    input_texts = [f"{prefix[mode]}{text}" for text in batch]
    inputs = tokenizer(input_texts, return_tensors="pt", padding=True, truncation=True, max_length=max_length).to(model.device)
    
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=max_length)
    
    return tokenizer.batch_decode(outputs, skip_special_tokens=True)
```

```python
def postprocess_sentence(sentence, mode="lemma-morphosyntax"):
    if mode == "lemma-morphosyntax" or mode == "segmentation-lemma-morphosyntax":
        result = ""
        for item in sentence.split(" "):
            if mode == "segmentation-lemma-morphosyntax":
                if len(item.split("_")) == 3:
                    unsandhied, lemma, short_tag = item.split("_")
                    if short_tag in sanskrit_tags:
                        short_tag = sanskrit_tags[short_tag]
                        if "Cpd" in short_tag:
                            unsandhied = unsandhied + "-"
                    result += f"{unsandhied}_{lemma}_{short_tag} "
            else:
                if len(item.split("_")) == 2:
                    lemma, short_tag = item.split("_")
                    if short_tag in sanskrit_tags:
                        short_tag = sanskrit_tags[short_tag]
                    result += f"{lemma}_{short_tag} "
        return result
    elif mode == "lemma" or mode == "segmentation":
        return sentence.replace("_", " ")
```

```python
def extract_pos(lemma, features):
    """
    Extracts the exact Universal Dependencies POS tag dynamically 
    using ONLY morphological feature flags and structural clues.
    """
    if not features or features == "UNK":
        return "UNKNOWN"

    # 1. PRONOUNS
    # Explicitly flagged or matches standard UD pronominal types
    if "PronType=" in features:
        return "PRON"

    # 2. VERBS & VERBAL FORMS (Infinitives, Gerunds, Participles)
    if "VerbForm=" in features or any(x in features for x in ["Tense=", "Mood=", "Person="]):
        # Infinitives (e.g., khāditum) and Finite Verbs (e.g., agacchat)
        if "VerbForm=Inf" in features or "Person=" in features:
            return "VERB"
        
        # Gerunds / Converb forms (e.g., āgatya, dṛṣṭvā) serve an adverbial role
        if "VerbForm=Conv" in features or "VerbForm=Ger" in features:
            return "ADV"
            
        # Past Passive/Active Participles (e.g., prasthitaḥ, kṛtavān) 
        # often act morphologically as adjectives or finite verbs.
        # If they carry case markers but lack explicit person, UD often treats them as ADJ/VERB.
        if "VerbForm=Part" in features:
            # If the features explicitly label it as an adjective variant
            if "Adj" in features:
                return "ADJ"
            return "VERB"

    # 3. NOMINALS (Nouns vs. Adjectives)
    if "Case=" in features:
        # Check if the feature string explicitly identifies it as an adjective or compound modifier
        if "Adj" in features or "Case=Cpd" in features:
            return "ADJ"
        return "NOUN"

    # 4. INVARIABLES / PARTICLES (Conjunctions, Adverbs, Adpositions)
    # Words without Case, Gender, Number, or Verb features are invariably particles
    if "Morf=Uninf" in features or not any(x in features for x in ["Case=", "Gender=", "Number="]):
        # If your mapping file or model maps sub-types:
        if "Adverb" in features:
            return "ADV"
        if "Conj" in features:
            return "CCONJ"
        
        # Default fallback for uninflected items (ca, iti, na, etc.)
        return "PART"

    return "UNKNOWN"
```

```python
def parse_feats(feats_str):
    feats = {}
    if not feats_str:
        return feats

    for item in feats_str.split("|"):
        if "=" in item:
            k, v = item.split("=", 1)
            feats[k] = v

    return feats
```

```python
mode = "segmentation-lemma-morphosyntax"
tokenized_transliterated_version=[]
# =====================================================================
# 3. RUN INFERENCE ON IAST SENTENCES WITH POS EXTRACTION
# =====================================================================

print("\nRunning inference...")
raw_outputs = process_batch(cleaned_transliterated_passage, mode=mode)
processed_outputs = [
    postprocess_sentence(out, mode=mode) for out in raw_outputs
]

print("\n=== Parsed Results ===")
for orig, parsed in zip(cleaned_transliterated_passage, processed_outputs):
    # print(f"\nOriginal (IAST): {orig}")
    # print("Token Breakdown:")

    for word_info in parsed.strip().split(" "):
        if not word_info.strip():
            continue
        cleaned_word_info = word_info.strip("_")
        parts = word_info.split("_")
        
        if len(parts) >= 2:
            token = parts[0]
            lemma = parts[1]
            features = parts[2] if (len(parts)>=3 and parts[2].strip())else "Morf=UnInf"
            
            pos = extract_pos(lemma, features)
            
            feat_dict = parse_feats(features)
            
            tokenized_transliterated_version.append({
                "text": token,
                "lemma":lemma,
                "upos":pos,
                "xpos":feat_dict.get("Gender", None),
                "feats":features                                                    
            })
            
            # print(
            #     f"  • Token: {token:<15} | Lemma: {lemma:<15} | POS: {primary_pos} | Full Tag: {features}"
            # )
        else:
            print(f"  • Raw Parsing Error/Format: {word_info}")
            
print(tokenized_transliterated_version)
```
