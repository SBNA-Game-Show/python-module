```python
import stanza
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate

# 1. Initialize Stanza (ensure 'vedic' package is loaded)
stanza.download("sa", package="vedic")
nlp = stanza.Pipeline(
    lang="sa", processors="tokenize,pos,lemma,depparse", package="vedic"
)

for sentence in sanskrit_passage:
    # Fix 1: Switch to SLP1 so Stanza doesn't misclassify tokens
    trans = transliterate(sentence, sanscript.DEVANAGARI, sanscript.HK)

    # Fix 2: Reassign back to the variable since strings are immutable
    trans = trans.replace("|", "")
    trans = trans.replace("-", "")
    trans = trans.strip()

    # 2. Tokenize using Stanza
    doc = nlp(trans)

    print(f"\nOriginal: {sentence}")
    print(f"Transliterated (SLP1): {trans}")
    print("-" * 40)

    # Print out Stanza's tokenization output
    for sent in doc.sentences:
        for word in sent.words:
            print(
                f"Token: {word.text:<12} | Lemma: {word.lemma:<12} | POS: {word.upos}"
            )
```
