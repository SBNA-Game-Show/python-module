```python
from utils.file_system_reader import FileSystemReader
```

```python
reader = FileSystemReader("Samskrutam_story.json")
raw_data = reader.read_file()
print(raw_data)
```

```python
sanskrit_passage = raw_data[0]["sanskritVersion"]
print(sanskrit_passage)
```

```python
cleaned_sanskrit_passage = []

for sentence in sanskrit_passage:
    sentence = sentence.replace("|", "")
    sentence = sentence.replace("||", "")
    sentence = sentence.replace("-", "")
    sentence = sentence.replace("–", "")
    sentence = sentence.replace("—", "")
    sentence = sentence.replace('"', "")
    sentence = sentence.replace("“", "")
    sentence = sentence.replace("”", "")
    sentence = sentence.replace(",", "")
    sentence = sentence.replace(".", "")
    sentence = sentence.replace("?", "")
    sentence = sentence.replace("!", "")
    sentence = sentence.replace(";", "")
    sentence = sentence.replace(":", "")
    sentence = sentence.replace("(", "")
    sentence = sentence.replace(")", "")

    cleaned_sanskrit_passage.append(sentence)
print(cleaned_sanskrit_passage)
```

```python
import requests
import uuid
import json
import re

url = "https://dharmamitra.org/bff/api/translation"

text = "\n".join(cleaned_sanskrit_passage)

payload = {
    "do_grammar": False,
    "input_encoding": "auto",
    "input_sentence": text,
    "messages": [
        {
            "parts": [
                {
                    "type": "text",
                    "text": text
                }
            ],
            "id": str(uuid.uuid4()),
            "role": "user"
        }
    ],
    "mode": "explain-grammar",
    "target_lang": "english"
}

headers = {
    "Accept": "text/event-stream",
    "Content-Type": "application/json",
    "Origin": "https://dharmamitra.org",
    "Referer": "https://dharmamitra.org/"
}

response = requests.post(
    url,
    json=payload,
    headers=headers,
    stream=True
)



full_text = ""

for line in response.iter_lines():
    if line:
        line = line.decode("utf-8")

        if line.startswith("data: "):
            data = line[6:]

            if data != "[DONE]":
                obj = json.loads(data)

                if obj.get("type") == "text-delta":
                    full_text += obj["delta"]


print(full_text)
```

```python
from utils.dharmamitra_parser import DharmamitraExtractor
ext = DharmamitraExtractor()
tok_data = ext.extract(full_text)
print(tok_data)
```

removing duplicate words

```python
cleaned_words_array = []
unique_words = set()

for item in tok_data:
    word = item["text"]

    if word not in unique_words:
        unique_words.add(word)
        cleaned_words_array.append(item)

print(json.dumps(cleaned_words_array, indent=2, ensure_ascii=False))
```

convert lemma to sanskrit

```python
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate
```

```python
final_array = []

for item in cleaned_words_array:  # Using the clean output from your extraction function
    text = item["text"]
    lemma = item["lemma"]
    
    # Transliterate the IAST lemma string directly to Devanagari
    text_devanagari = sanscript.transliterate(text, sanscript.IAST, sanscript.DEVANAGARI)
    lemma_devanagari = sanscript.transliterate(lemma, sanscript.IAST, sanscript.DEVANAGARI)
    
    # Construct the final entry with both representations
    final_array.append({
        "text": text_devanagari,
        "lemma": lemma_devanagari,
        "upos": item["upos"],
        "features": item["features"],
        "definition": item["definition"]
    })

# Output the fully structured array
print(json.dumps(final_array, indent=2, ensure_ascii=False))
```
