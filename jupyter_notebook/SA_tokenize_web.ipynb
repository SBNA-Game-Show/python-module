```python
reader = FileSystemReader("Samskrutam_story.json")
raw_data = reader.read_file()
print(raw_data)
```

```python
sanskrit_passage = raw_data[0]["sanskritVersion"]
print(sanskrit_passage)
```

```python
cleaned_sanskrit_passage = []

for sentence in sanskrit_passage:
    sentence = sentence.replace("|", "")
    sentence = sentence.replace("||", "")
    sentence = sentence.replace("-", "")
    sentence = sentence.replace("–", "")
    sentence = sentence.replace("—", "")
    sentence = sentence.replace('"', "")
    sentence = sentence.replace("“", "")
    sentence = sentence.replace("”", "")
    sentence = sentence.replace(",", "")
    sentence = sentence.replace(".", "")
    sentence = sentence.replace("?", "")
    sentence = sentence.replace("!", "")
    sentence = sentence.replace(";", "")
    sentence = sentence.replace(":", "")
    sentence = sentence.replace("(", "")
    sentence = sentence.replace(")", "")

    cleaned_sanskrit_passage.append(sentence)
print(cleaned_sanskrit_passage)
```

```python
import requests
import uuid
import json
import re

url = "https://dharmamitra.org/bff/api/translation"

text = "\n".join(cleaned_sanskrit_passage)

payload = {
    "do_grammar": False,
    "input_encoding": "auto",
    "input_sentence": text,
    "messages": [
        {
            "parts": [
                {
                    "type": "text",
                    "text": text
                }
            ],
            "id": str(uuid.uuid4()),
            "role": "user"
        }
    ],
    "mode": "explain-grammar",
    "target_lang": "english"
}

headers = {
    "Accept": "text/event-stream",
    "Content-Type": "application/json",
    "Origin": "https://dharmamitra.org",
    "Referer": "https://dharmamitra.org/"
}

response = requests.post(
    url,
    json=payload,
    headers=headers,
    stream=True
)



full_text = ""

for line in response.iter_lines():
    if line:
        line = line.decode("utf-8")

        if line.startswith("data: "):
            data = line[6:]

            if data != "[DONE]":
                obj = json.loads(data)

                if obj.get("type") == "text-delta":
                    full_text += obj["delta"]


print(full_text)
```

```python
import re
import json


def extract_dharmamitra_words(full_text):
    results = []

    # Regex matching the Markdown formatting accurately
    pattern = re.compile(
        r"\*\s+\*\*(.*?)\*\*\s+\((.*?)\),\s+(.*?),\s+_(.*?)_",
        re.DOTALL
    )

    # Dictionary to map Dharmamitra labels to valid UD UPOS standards
    pos_mapping = {
        "pronoun": "PRON",
        "particle": "PART",
        "noun": "NOUN",
        "compound noun": "NOUN",
        "verb": "VERB",
        "adjective": "ADJ",
        "adverb": "ADV"
    }

    for word, lemma, grammar, definition in pattern.findall(full_text):
        word = word.strip()
        lemma = lemma.strip()
        grammar = grammar.strip()
        definition = definition.strip()

        # Split the grammatical elements safely
        grammar_parts = [x.strip() for x in grammar.split(",")]
        
        if grammar_parts:
            raw_pos = grammar_parts[0].lower()
            # Map to standard UD UPOS labels, fallback to raw string if missing
            upos = pos_mapping.get(raw_pos, raw_pos.upper())
            # Remaining elements are morphological features
            features_list = grammar_parts[1:]
        else:
            upos = "X"
            features_list = []

        results.append({
            "text": word,
            "lemma": lemma,
            "upos": upos,
            "features": ", ".join(features_list),
            "definition": definition
        })

    return results


words_array = extract_dharmamitra_words(full_text)

print(json.dumps(words_array,
                 indent=2,
                 ensure_ascii=False))
```

removing duplicate words

```python
cleaned_words_array = []
unique_words = set()

for item in words_array:
    word = item["text"]

    if word not in unique_words:
        unique_words.add(word)
        cleaned_words_array.append(item)

print(json.dumps(cleaned_words_array, indent=2, ensure_ascii=False))
```

convert lemma to sanskrit

```python
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate
```

```python
final_array = []

for item in words_array:  # Using the clean output from your extraction function
    lemma = item["lemma"]
    
    # Transliterate the IAST lemma string directly to Devanagari
    lemma_devanagari = sanscript.transliterate(lemma, sanscript.IAST, sanscript.DEVANAGARI)
    
    # Construct the final entry with both representations
    final_array.append({
        "text": item["text"],
        "lemma": lemma_devanagari,
        "upos": item["upos"],
        "features": item["features"],
        "definition": item["definition"]
    })

# Output the fully structured array
print(json.dumps(final_array, indent=2, ensure_ascii=False))
```
