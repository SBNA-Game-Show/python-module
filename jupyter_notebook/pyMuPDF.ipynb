```python
import pymupdf
import pytesseract
from PIL import Image
import numpy as np

doc = pymupdf.open(file_path)
page = doc[0]

data = page.get_text("dict")

for block in data["blocks"]:
    if "lines" not in block:
        continue

    for line in block["lines"]:
        for span in line["spans"]:
            if "NirmalaUI" not in span["font"]:
                continue

            rect = pymupdf.Rect(span["bbox"])

            # zoom for OCR quality
            matrix = pymupdf.Matrix(4, 4)

            pix = page.get_pixmap(
                matrix=matrix,
                clip=rect
            )

            img = Image.fromarray(
                np.frombuffer(
                    pix.samples,
                    dtype=np.uint8
                ).reshape(
                    pix.height,
                    pix.width,
                    pix.n
                )
            )

            text = pytesseract.image_to_string(
                img,
                lang="san"
            )

            print("PDF:", span["text"])
            print("OCR:", text.strip())
            print("-" * 40)
```

```python
import pymupdf
import pytesseract
import numpy as np
from PIL import Image


def ocr_sanskrit_block(page, zoom=6):
    """
    Find all NirmalaUI Sanskrit spans and OCR them as one block.
    """
    data = page.get_text("dict")

    sanskrit_spans = []

    for block in data["blocks"]:
        if "lines" not in block:
            continue

        for line in block["lines"]:
            for span in line["spans"]:
                if "NirmalaUI" in span["font"]:
                    sanskrit_spans.append(span)

    if not sanskrit_spans:
        return ""

    # Bounding box covering all Sanskrit text
    x0 = min(span["bbox"][0] for span in sanskrit_spans)
    y0 = min(span["bbox"][1] for span in sanskrit_spans)
    x1 = max(span["bbox"][2] for span in sanskrit_spans)
    y1 = max(span["bbox"][3] for span in sanskrit_spans)

    rect = pymupdf.Rect(x0, y0, x1, y1)

    pix = page.get_pixmap(
        matrix=pymupdf.Matrix(zoom, zoom),
        clip=rect,
        alpha=False
    )

    img = Image.fromarray(
        np.frombuffer(
            pix.samples,
            dtype=np.uint8
        ).reshape(
            pix.height,
            pix.width,
            pix.n
        )
    )

    # OCR Sanskrit block
    text = pytesseract.image_to_string(
        img,
        lang="san",          # use "hin" if san isn't installed
        config="--psm 6"
    )

    return text.strip()


def extract_english(page):
    """
    Extract only non-NirmalaUI text from PDF.
    """
    data = page.get_text("dict")

    english_parts = []

    for block in data["blocks"]:
        if "lines" not in block:
            continue

        for line in block["lines"]:
            line_text = []

            for span in line["spans"]:
                if "NirmalaUI" not in span["font"]:
                    line_text.append(span["text"])

            if line_text:
                english_parts.append("".join(line_text))

    return "\n".join(english_parts)


def process_pdf(file_path):
    doc = pymupdf.open(file_path)

    for page_num, page in enumerate(doc, start=1):
        print("=" * 80)
        print(f"PAGE {page_num}")
        print("=" * 80)

        english_text = extract_english(page)
        sanskrit_text = ocr_sanskrit_block(page)

        if english_text.strip():
            print("\nENGLISH:\n")
            print(english_text)

        if sanskrit_text.strip():
            print("\nSANSKRIT (OCR):\n")
            print(sanskrit_text)

    doc.close()


# Usage
process_pdf(file_path)
```
