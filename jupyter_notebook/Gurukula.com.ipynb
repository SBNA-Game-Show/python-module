```python
import requests
import urllib3
```

```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
base_url = "https://gurukula.com/sa"
```

```python
base_response = requests.get(base_url,timeout=10)

print(base_response.text)

base_response_data = base_response.text
```

```python
from bs4 import BeautifulSoup
```

```python
MASTER_DATA=[]
```

```python
soup = BeautifulSoup(base_response_data,"html.parser")
```

```python
raw_category_tags = soup.find_all("div",class_="all-epics-card")
print(raw_category_tags)
```

```python
categories = []

for card in raw_category_tags:
    link = card.find("a", href=True)
    name = card.find("div", class_="epic-name")

    if link and name:
        categories.append({
            "category": name.get_text(strip=True),
            "category_link": link["href"]
        })

print(categories)
```

Fetching story titles from categories

```python
cat_url = "https://gurukula.com/sa/saralakathah"
cat_req = requests.get(cat_url,timeout=10)
cat_response = cat_req.text
print(cat_response)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(cat_response, "html.parser")

comp_list = soup.find_all("div", class_="box")

passage_list = []

for box in comp_list:
    link = box.select_one(".heading a")

    if link:
        passage_list.append({
            "url": link.get("href"),
            "text": link.get_text(strip=True)
        })

print(passage_list)
```

```python
print(len(passage_list))
```

```python
master_data = []

for category in categories:
    master_data.append({
        "category": category["category"],
        "category_link": category["category_link"],
        "passages": []
    })


for passage in passage_list:
    passage_url = passage["url"]

    for category in master_data:
        category_url = category["category_link"]

        # extract category path and compare
        if passage_url.startswith(category_url):
            category["passages"].append(passage)
            break


print(master_data)
```
