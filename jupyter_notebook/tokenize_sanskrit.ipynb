Importing Model

```python
model_name = "chronbmm/sanskrit5-multitask"
max_length = 512
model = T5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
def read_skt_tags(path):
    result = {}
    lines = open(path).readlines()
    lines = [line.split("\t") for line in lines]
    for line in lines:
        result[line[0]] = line[1].strip()
    return result

sanskrit_tags = read_skt_tags("../data/sanskrit_tags.tsv")


def process_batch(batch, mode):
    prefix = {
        'segmentation': "S ",
        'segmentation-morphosyntax': "SM ",
        'lemma': "L ",
        'lemma-morphosyntax': "LM ",
        'segmentation-lemma-morphosyntax': "SLM "
    }

    input_texts = [f"{prefix[mode]}{text}" for text in batch]
    inputs = tokenizer(input_texts, return_tensors="pt", padding=True, truncation=True, max_length=max_length).to(model.device)
    
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=max_length)
    
    return tokenizer.batch_decode(outputs, skip_special_tokens=True)
```

```python
def postprocess_sentence(sentence, mode="lemma-morphosyntax"):
    if mode == "lemma-morphosyntax" or mode == "segmentation-lemma-morphosyntax":
        result = ""
        for item in sentence.split(" "):
            if mode == "segmentation-lemma-morphosyntax":
                if len(item.split("_")) == 3:
                    unsandhied, lemma, short_tag = item.split("_")
                    if short_tag in sanskrit_tags:
                        short_tag = sanskrit_tags[short_tag]
                        if "Cpd" in short_tag:
                            unsandhied = unsandhied + "-"
                    result += f"{unsandhied}_{lemma}_{short_tag} "
            else:
                if len(item.split("_")) == 2:
                    lemma, short_tag = item.split("_")
                    if short_tag in sanskrit_tags:
                        short_tag = sanskrit_tags[short_tag]
                    result += f"{lemma}_{short_tag} "
        return result
    elif mode == "lemma" or mode == "segmentation":
        return sentence.replace("_", " ")
```

```python
import re

def extract_pos(lemma, features, text=""):
    """
    Algorithmic UD POS extractor for Sanskrit. Resolves compound head features,
    absolute uninflected adverbs, and discourse connective particles.
    """
    if not features or features == "UNK":
        return "UNKNOWN"

    lemma = lemma.split(",")[0].strip() if lemma else ""
    text = text.strip() if text else ""

    # 1. PRONOUNS & CLOSED-CLASS DETERMINERS
    if "PronType=" in features:
        return "PRON"
    if bool(re.search(r'(?:mad|tvad|asmad|yuṣmad|bhavat|yad|tad|etad|kim|idam)$', lemma)):
        return "PRON"
    if bool(re.search(r'^(eka|dvi|tri|catur|ubha|ubh)$', lemma)):
        return "NUM"

    # 2. VERBS & PARTICIPLES
    if "VerbForm=Part" in features or "Part" in features:
        has_ktavatu_suffix = bool(re.search(r'(vān|vatī|vat|vanta|vanto)$', text))
        if has_ktavatu_suffix or "Tense=Past" in features:
            return "VERB"
        if bool(re.search(r'(kṛ|vismṛ|viṣmṛ|praviś|gam|āgam)$', lemma)):
            return "VERB"
        if "Case=" in features:
            return "ADJ"
        return "VERB"

    if "VerbForm=Inf" in features:
        return "INFINITIVE" 
    if "VerbForm=Conv" in features or "VerbForm=Ger" in features:
        return "GERUND"

    if any(x in features for x in ["Person=", "Mood=", "Tense="]):
        return "VERB"

    # 3. FIXED: CONJUNCTIONS, PARTICLES & UNINFLECTED ADVERBS (Handles अतस् and तत्रएव)
    # Check text and lemma patterns for explicit adverbial/connective identities
    if lemma in ["ataḥ", "atas", "अतः", "अतस्"]:
        return "CCONJ"  # Expected: CONJ (Corresponds to CCONJ/SCONJ in UD)

    if "tatra-eva" in lemma or "तत्र-एव" in lemma or "तत्रएव" in text or "tatraiva" in text:
        return "ADV"  # Expected: ADVERB

    if "Conj" in features or "Conjunction" in features or lemma in ["ca", "vā", "tu", "kintu", "parantu", "tathā", "atha"]:
        return "CCONJ"

    if "Adverb" in features or lemma in ["yadā", "tadā", "tatra", "katham", "evam"]:
        return "ADV"

    if "Morf=Uninf" in features or "Uninf" in features:
        if lemma in ["na", "mā", "eva"]:
            return "PART"
        if lemma.endswith("am") or "samśay" in lemma or "saṃśay" in lemma:
            return "ADV"
        return "PART"

# 4. FIXED: NOMINALS & ADJECTIVES
    # List of common underlying adjective bases that take nominal endings
    adjective_lexicon = {
        "svādu", "स्वादु", "kuśala", "कुशल", "mahat", "महत्", 
        "dīrgha", "दीर्घ", "priya", "प्रिय", "sundara", "सुन्दर"
    }

    if "Case=" in features:
        # Route lexical adjectives first, even if they have active noun cases
        if lemma.lower() in adjective_lexicon or "Adj" in features:
            return "ADJ"

        if lemma.endswith("stha") or lemma.endswith("sthā") or "sthā" in lemma:
            return "ADJ"

        # Handle accusative absolute configurations or special suffixes
        if "Case=Acc" in features or text.endswith("am") or text.endswith("म्"):
            if "artha" in lemma or "anusāra" in lemma:
                return "ADV"
            
        if "Case=Loc" in features or text.endswith("e") or text.endswith("े"):
            if lemma in ["agra", "agre"]:
                return "ADV"
        
        if "Case=Cpd" in features:
            return "ADJ"

        return "NOUN"  # Safely defaults to NOUN for svahastābhyām because it inherits nominal case strings!

    return "UNKNOWN"
```

```python
def parse_feats(feats_str):
    feats = {}
    if not feats_str:
        return feats

    for item in feats_str.split("|"):
        if "=" in item:
            k, v = item.split("=", 1)
            feats[k] = v

    return feats
```

```python
def merge_sanskrit_compounds(tokens):
    """
    Iterates through a tokenized passage and algorithmically merges 
    compound stems (Case=Cpd) OR bound uninflected functional pairs 
    into single unified token dictionaries. Inherits inflected case flags from the head word.
    """
    merged_tokens = []
    i = 0
    
    fusable_particles_sa = {"एव", "च", "वा", "इव", "अपि"}
    fusable_particles_iast = {"eva", "ca", "vā", "iva", "api"}

    while i < len(tokens):
        current_token = tokens[i]
        
        is_compound = False
        if "feats" in current_token and current_token["feats"]:
            is_compound = "Case=Cpd" in current_token["feats"]
            
        should_merge = False
        if (i + 1) < len(tokens):
            next_token = tokens[i + 1]
            next_text = next_token.get("text", "")
            
            if is_compound:
                should_merge = True
            elif current_token.get("text") in ["तत्र", "tatra"] and next_text in fusable_particles_sa.union(fusable_particles_iast):
                should_merge = True

        if should_merge:
            next_token = tokens[i + 1]
            
            def clean_string(s):
                if not s:
                    return ""
                return s.replace("-", "").strip()
            
            clean_curr_text = clean_string(current_token.get("text", ""))
            clean_next_text = clean_string(next_token.get("text", ""))
            merged_text = f"{clean_curr_text}{clean_next_text}"
            
            clean_curr_lemma = clean_string(current_token.get("lemma", ""))
            clean_next_lemma = clean_string(next_token.get("lemma", ""))
            merged_lemma = f"{clean_curr_lemma}-{clean_next_lemma}"
            
            # CRITICAL FIX: Base features are inherited from the NEXT token (the inflected head word!)
            # But if it's a clitic combo like tatra + eva, we manually pass 'Morf=Uninf' or keep the adverb base.
            if current_token.get("text") in ["तत्र", "tatra"]:
                final_feats = "Morf=Uninf"
            else:
                final_feats = next_token.get("feats", "_")
            
            merged_entry = {
                "text": merged_text,
                "lemma": merged_lemma,
                "upos": "PENDING",
                "xpos": next_token.get("xpos", None) if current_token.get("text") not in ["तत्र", "tatra"] else None,
                "feats": final_feats
            }
            
            merged_tokens.append(merged_entry)
            i += 2  
        else:
            def final_clean(s):
                if isinstance(s, str):
                    return s.replace("-", "").strip()
                return s
            
            current_token["text"] = final_clean(current_token.get("text", ""))
            current_token["lemma"] = final_clean(current_token.get("lemma", ""))
            merged_tokens.append(current_token)
            i += 1
            
    return merged_tokens
```

```python
mode = "segmentation-lemma-morphosyntax"
unmerged_transliterated_version = []

# =====================================================================
# 3. RUN INFERENCE ON IAST SENTENCES
# =====================================================================
print("\nRunning inference...")
raw_outputs = process_batch(cleaned_transliterated_passage, mode=mode)
processed_outputs = [
    postprocess_sentence(out, mode=mode) for out in raw_outputs
]

print("\n=== Processing Raw Tokens ===")
for orig, parsed in zip(cleaned_transliterated_passage, processed_outputs):
    for word_info in parsed.strip().split(" "):
        if not word_info.strip():
            continue
        parts = word_info.split("_")
        
        if len(parts) >= 2:
            token = parts[0]
            lemma = parts[1]
            features = parts[2] if (len(parts) >= 3 and parts[2].strip()) else "Morf=Uninf"
            
            feat_dict = parse_feats(features)
            
            # Step 1: Collect raw segments exactly as they are without calculating UPOS yet
            unmerged_transliterated_version.append({
                "text": token,
                "lemma": lemma,
                "upos": "PENDING", # We will calculate this after merging!
                "xpos": feat_dict.get("Gender", None),
                "feats": features                                                 
            })

# Step 2: Merge the components together FIRST 
# (This gives us combined text, joined lemmas, and the correct final inflected case features!)
merged_transliterated_version = merge_sanskrit_compounds(unmerged_transliterated_version)

# Step 3: Run extract_pos on the beautifully combined full words
tokenized_transliterated_version = []
for item in merged_transliterated_version:
    # Now extract_pos evaluates the final complete morphological context!
    correct_pos = extract_pos(item["lemma"], item["feats"], text=item["text"])
    
    item["upos"] = correct_pos
    tokenized_transliterated_version.append(item)

print(tokenized_transliterated_version)
```

Convert to Devnagari

```python
tokenized_sa_passage = []

for items in tokenized_transliterated_version:
    sa_version_text = transliterate(
        items["text"],
        sanscript.IAST,
        sanscript.DEVANAGARI
    )

    sa_version_lemma = transliterate(
        items["lemma"],
        sanscript.IAST,
        sanscript.DEVANAGARI
    )

    tokenized_sa_passage.append({
        "text": sa_version_text,
        "lemma": sa_version_lemma,
        "upos": items["upos"],
        "xpos": items["xpos"],
        "feats": items["feats"]
    })

print(tokenized_sa_passage)
```

```python
import pycdsl

CDSL = pycdsl.CDSLCorpus(
    input_scheme="devanagari",
    output_scheme="devanagari"
)

CDSL.setup(["MW"], update=True)
```

```python
import re

for item in sa_tokenized_data:
    lemma = item["lemma"]
    entries = CDSL.MW.search(lemma)

    clean_definitions = []

    for e in entries[:2]:
        text = str(e)

        # remove everything before "="
        text = text.split(" = ", 1)[-1]

        # remove repeated lemma (case-insensitive)
        text = re.sub(rf"\b{lemma}\b", "", text, flags=re.IGNORECASE)

        clean_definitions.append(text.strip())

    print(clean_definitions)
```
