```python
import pymupdf
import pytesseract
from PIL import Image
import numpy as np
import re
file_path = os.path.join("..","data","sanskrit_story_single.pdf")
print(file_path)
```

```python
doc = pymupdf.open(file_path)

for page_num, page in enumerate(doc):
    pix = page.get_pixmap(matrix=pymupdf.Matrix(4, 4))

    img = Image.fromarray(
        np.frombuffer(
            pix.samples,
            dtype=np.uint8
        ).reshape(
            pix.height,
            pix.width,
            pix.n
        )
    )

    text = pytesseract.image_to_string(
        img,
        lang="san"
    )
    lines =[line.strip() for line in text.splitlines() if line.strip()]
    title = re.sub(r'^[०-९]+\.\s*', '', lines[0])
    body = " ".join(lines[1:])
       
    print(title)
    print(body)
```

Reading Multiple sanskrit passages from PDF

```python
import pymupdf
import pytesseract
from PIL import Image
import numpy as np
import re
file_path = os.path.join("..","data","sanskrit_stories_five.pdf")
print(file_path)
```

```python
doc = pymupdf.open(file_path)

for page_num, page in enumerate(doc):
    pix = page.get_pixmap(matrix=pymupdf.Matrix(4, 4))

    img = Image.fromarray(
        np.frombuffer(
            pix.samples,
            dtype=np.uint8
        ).reshape(
            pix.height,
            pix.width,
            pix.n
        )
    )

    text = pytesseract.image_to_string(
        img,
        lang="san"
    )
    lines =[line.strip() for line in text.splitlines() if line.strip()]
    title = re.sub(r'^[०-९]+\.\s*', '', lines[0])
    body = " ".join(lines[1:])
       
    print(title)
    print(body)
```

```python
passage_array = []
current_title = None
current_text = []

doc = pymupdf.open(file_path)

for page in doc:
    pix = page.get_pixmap(matrix=pymupdf.Matrix(4, 4))

    img = Image.fromarray(
        np.frombuffer(pix.samples, dtype=np.uint8)
        .reshape(pix.height, pix.width, pix.n)
    )

    text = pytesseract.image_to_string(img, lang="san")

    lines = text.splitlines()

    for i in range(len(lines)):
        line = lines[i].strip()

        prev_empty = (i == 0) or (lines[i - 1].strip() == "")
        next_empty = (i == len(lines) - 1) or (lines[i + 1].strip() == "")

        # TITLE RULE: surrounded by blank lines
        if line and prev_empty and next_empty:

            # save previous passage
            if current_title and current_text:
                passage_array.append({
                    "title": current_title,
                    "text": " ".join(current_text)
                })

            current_title = line
            current_text = []

        else:
            if current_title and line:
                current_text.append(line)

# save last passage
if current_title and current_text:
    passage_array.append({
        "title": current_title,
        "text": " ".join(current_text)
    })
```

```python
print(passage_array)
```
