```python
import requests
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
init_url = "https://gurukula.com/sa"
base_url = "https://gurukula.com"
MASTER_DATA ={}
```

Extract all available categories

```python
category_query = requests.get(init_url,timeout=10)
category_query_raw_data = category_query.text
print(category_query_raw_data)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(category_query_raw_data,"html.parser")
```

```python
# Extracting only category block
raw_category_tags = soup.find_all("div",class_="all-epics-card")
print(raw_category_tags)
```

```python
# Extracting categories and their url
categories = []

for card in raw_category_tags:
    link = card.find("a", href=True)
    name = card.find("div", class_="epic-name")

    if link and name:
        categories.append({
            "category": name.get_text(strip=True),
            "category_link": link["href"]
        })

print(categories)
```

Extracting story titles

```python
story_list=[]
for category in categories:
    link = category["category_link"]
    base_link = base_url+link
    response = requests.get(base_link,timeout=10)
    cat_data_response = response.text
    soup = BeautifulSoup(cat_data_response,"html.parser")
    raw_tags = soup.find_all("div",class_="box")
    for tags in raw_tags:
        link = tags.select_one(".heading a")

        if link:
            story_list.append({
                "url":link.get("href"),
                "text":link.get_text(strip=True)
            })
print(story_list)
```

writing to master file

```python
for category in categories:
    category_name = category["category"]
    category_link = category["category_link"]

    MASTER_DATA[category_name] = {
        "category_link": category_link,
        "passages": []
    }

    for story in story_list:
        if story["url"].startswith(category_link + "/"):
            MASTER_DATA[category_name]["passages"].append(story)
```

clean MetaData

```python
import re

for category_name, category_data in MASTER_DATA.items():
    for passage in category_data["passages"]:
        passage["text"] = re.sub(r"^\d+\.", "", passage["text"]).strip()

print(MASTER_DATA)
```
