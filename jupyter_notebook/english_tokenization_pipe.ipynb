```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

Tokenizing Single Word

```python
ENGLISH_WORD = "rabbit"
```

```python
doc = nlp(ENGLISH_WORD)
```

```python
tokens = []

for token in doc:
    tokens.append({
        "text": token.text,
        "lemma": token.lemma_,
        "pos": token.pos_,
        "tag": token.tag_
    })

print(tokens)
```

```python
import nltk
nltk.download("wordnet")
nltk.download("omw-1.4")
from nltk.corpus import wordnet
```

```python
synonyms = []

for syn in wordnet.synsets(ENGLISH_WORD):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())

print(set(synonyms))
```

```python
antonyms = []

for syn in wordnet.synsets(ENGLISH_WORD):
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

print(set(antonyms))
```

```python
pos_map = {
    "NOUN": wordnet.NOUN,
    "VERB": wordnet.VERB,
    "ADJ": wordnet.ADJ,
    "ADV": wordnet.ADV,
}

synonyms = set()

for syn in wordnet.synsets(
    tokens[0]["lemma"],
    pos=pos_map.get(tokens[0]["pos"])
):
    synonyms.update(syn.lemma_names())

print(synonyms)
```

```python
antonyms = set()

for syn in wordnet.synsets(
    tokens[0]["lemma"],
    pos=pos_map.get(tokens[0]["pos"])
):
    for lemma in syn.lemmas():
        for antonym in lemma.antonyms():
            antonyms.add(antonym.name())

print(antonyms)
```

Tokenizing Pipe line

```python
from utils.file_system_reader import FileSystemReader
```

```python
request = FileSystemReader("stories_raw.json")
RAW_DATA = request.read_file()
print(RAW_DATA)
```

```python
ENGLISH_VERSION_STORY = RAW_DATA["englishVersion"]
print(ENGLISH_VERSION_STORY)
```

```python
ENGLISH_VERSION_STORY_ARRAY = ENGLISH_VERSION_STORY.split()
print(ENGLISH_VERSION_STORY_ARRAY)
```

```python
TOKENIZED_ENGLISH_WORDS=[]

for word in ENGLISH_VERSION_STORY_ARRAY:
    doc = nlp(word)
    for token in doc:
        TOKENIZED_ENGLISH_WORDS.append({
            "text": token.text,
            "lemma": token.lemma_,
            "pos": token.pos_,
            "tag": token.tag_
        })

print(TOKENIZED_ENGLISH_WORDS)
```

```python

SYNONYMIZED_ENGLISH_WORDS = []

for token in TOKENIZED_ENGLISH_WORDS:
    synonyms = set()

    for syn in wordnet.synsets(
        token["lemma"],
        pos=pos_map.get(token["pos"])
    ):
        synonyms.update(syn.lemma_names())

    SYNONYMIZED_ENGLISH_WORDS.append({
        "text": token["text"],
        "lemma": token["lemma"],
        "pos": token["pos"],
        "synonyms": sorted(list(synonyms))
    })

print(SYNONYMIZED_ENGLISH_WORDS)
```

```python
ANTONYMISED_ENGLISH_WORDS = []

for token in SYNONYMIZED_ENGLISH_WORDS:
    antonyms = set()

    for syn in wordnet.synsets(
        token["lemma"],
        pos=pos_map.get(token["pos"])
    ):
        for lemma in syn.lemmas():
            for antonym in lemma.antonyms():
                antonyms.add(antonym.name())

    ANTONYMISED_ENGLISH_WORDS.append({
        "text": token["text"],
        "lemma": token["lemma"],
        "pos": token["pos"],
        "antonyms": sorted(list(antonyms))
    })

print(ANTONYMISED_ENGLISH_WORDS)
```

```python
antonym_lookup = {
    item["lemma"]: item["antonyms"]
    for item in ANTONYMISED_ENGLISH_WORDS
}

antonym_lookup
```

```python
ENRICHED_WORDS = []

for word in SYNONYMIZED_ENGLISH_WORDS:
    ENRICHED_WORDS.append({
        **word,
        "antonyms": antonym_lookup.get(
            word["lemma"],
            []
        )
    })

ENRICHED_WORDS
```
