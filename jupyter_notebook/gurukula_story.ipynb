```python
import requests
import urllib3
import re
from bs4 import BeautifulSoup
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
base_url = "https://gurukula.com/sa/tenali-raman/the-boon-of-kali-devi-2/story"
```

```python
base_response = requests.get(base_url,timeout=10)
raw_html = base_response.text
print(raw_html)
```

Parse HTML

```python
print(type(raw_html))
```

```python
from bs4 import BeautifulSoup
import json

soup = BeautifulSoup(raw_html, "html.parser")

next_script = soup.find("script", id="__NEXT_DATA__")

if next_script:
    raw_json = next_script.string
    next_data = json.loads(raw_json)

    print(type(next_data))
else:
    print("No __NEXT_DATA__ found")
```

```python
page_data = next_data["props"]["pageProps"]["data"]

print(page_data.keys())
```

```python
buffer_data = page_data["storyBufferData"]["Story"]["data"]

story_html = bytes(buffer_data).decode("utf-8")

print(story_html)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(story_html, "html.parser")

paragraphs = soup.find_all("p")

sanskrit_text = []

for p in paragraphs:
    text = p.get_text(" ", strip=True)

    if text:
        sanskrit_text.append(text)

story = "\n".join(sanskrit_text)

print(story)
```

```python
paragraphs = soup.find_all("p")

sanskrit_paragraphs = []

for p in paragraphs:
    text = p.get_text(" ", strip=True)

    # skip empty paragraphs
    if not text:
        continue

    # skip translator note
    if text.startswith("Translated"):
        continue

    sanskrit_paragraphs.append(text)


sanskrit_text = "\n".join(sanskrit_paragraphs)

print(sanskrit_text)
```
