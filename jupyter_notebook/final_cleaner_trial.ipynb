```python
from utils.file_system_reader import FileSystemReader
FILE_NAME = "tokenized_stories.json"
```

```python
reader = FileSystemReader(FILE_NAME)
data = reader.read_file()
print(data)
```

```python
ENGLISH_ARRAY = data[0]["tokenized_english_version"]
print(ENGLISH_ARRAY)
```

```python
english_array_copy = []

for item in ENGLISH_ARRAY:
    # 1. Skip spaces and punctuation completely
    if item.get("pos") in ["SPACE", "PUNCT"]:
        continue  
        
    # 2. Copy and append the valid items
    cleaned_item = item.copy()
    english_array_copy.append(cleaned_item)

print(english_array_copy)
```

```python
for item in english_array_copy:
    print(item)
```

```python
english_array_copy_final = []

for item in english_array_copy:
    # 1. Create a copy of the dictionary
    cleaned_item = item.copy()
    
    # 2. Find all keys where the value is an empty list
    empty_keys = [key for key, value in cleaned_item.items() if value == []]
    
    # 3. Delete those keys from the cleaned item
    for key in empty_keys:
        cleaned_item.pop(key)
        
    # 4. Append the cleaned dictionary to the copy array
    english_array_copy_final.append(cleaned_item)

print(english_array_copy_final)
```

```python
print(len(english_array_copy_final))
```

```python
def remove_duplicate_tokens(tokens):
    seen = set()
    result = []

    for item in tokens:
        word = item.get("text")
        if not word:
            continue

        key = word.lower().strip()

        if key in seen:
            continue

        seen.add(key)
        result.append(item)

    return result
```

```python
cleaned_tokens = remove_duplicate_tokens(english_array_copy_final)

print(cleaned_tokens)
```

```python
print(len(cleaned_tokens))
```
