```python
import pycdsl

CDSL = pycdsl.CDSLCorpus(
    input_scheme="devanagari",
    output_scheme="devanagari"
)

CDSL.setup(["MW"], update=True)
```

```python
print(CDSL.dicts.keys())
```

```python
results = CDSL.MW.search("शशक")
print(results)
```

```python
for entry in results:
    print(entry)
```

```python
results = CDSL.MW.search("शशक")

for r in results:
    text = str(r)
    headword = text.split(":")[2].split("=")[0].strip()
    print(headword)
```

```python
for r in CDSL.MW.search("नकुल"):
    text = str(r)

    meaning = text.split("=", 1)[1].strip()
    print(meaning)
```
